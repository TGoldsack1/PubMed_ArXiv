## Function Definition

```python
''' Declare function needed later '''

def conflate_headings(heading_text):
    if "conclu" in heading_text.lower() or "summary" in heading_text.lower():
        return "[conclusion]"
    if "introduction" in heading_text.lower():
        return "[introduction]"
    if "result" in heading_text.lower() or "analys" in heading_text.lower() or "ablat" in heading_text.lower():
        return "[results/analysis]"
    if "discussion" in heading_text.lower():
        return "[discussion]"
    if "acknowledg" in heading_text.lower():
        return "[acknowledgments]"
    if "background" in heading_text.lower() or "motivation" in heading_text.lower():
        return "[background/motivation]"
    if "related" in heading_text.lower():
        return "[related_work]"
    if "experiment" in heading_text.lower():
        return "[experiments]"
    if "method" in heading_text.lower() or "implement" in heading_text.lower():
        return "[method]"
    if "evaluation" in heading_text.lower():
        return "[evaluation]"
    if "model" in heading_text.lower() or "architec" in heading_text.lower() or "system" in heading_text.lower():
        return "[model]"
    if "data" in heading_text.lower():
        return "[data]"
    if "future" in heading_text.lower() or "direction" in heading_text.lower():
        return "[future_work]"
    if "case report" in heading_text.lower():
        return "[case report]"
    if "images" in heading_text.lower():
        return "[images]"
    
    return heading_text
```

### N Most Similar 
For each abstract sentence, look at the N most similar instances from the text + where they are located.

```python
''' Go through each instance and retrieve the most similar sentence from each section '''

N = 1

def get_most_similar_df(ds_dict, keys, n):
    most_similar_lines_flat = []
    most_similar_intro_lines_flat = []
    most_similar_conc_lines_flat = []

    
    for key in list(keys):
        inst = ds_dict[key]
        
        most_similar_intro_lines_flat = most_similar_intro_lines_flat + \
            [x["most_similar_intro_line"] if "most_similar_intro_line" in x.keys() else None for x in inst]
        
        most_similar_conc_lines_flat = most_similar_conc_lines_flat + \
            [x["most_similar_conc_line"] if "most_similar_conc_line" in x.keys() else None for x in inst]
        
        for i in range(n):
            most_similar_lines = [x[f"most_similar_line_{i}"] for x in inst]
            most_similar_lines_flat = most_similar_lines_flat + most_similar_lines


    df = pd.DataFrame()
    df['abstract_sentence_idx'] = [x['abstract_sentence_idx'] for x in most_similar_lines_flat]
    df['section_name'] = [x['section_name'] for x in most_similar_lines_flat]
    df['sentence_idx'] = [x['sentence_idx'] for x in most_similar_lines_flat]
    df['sentence_similarity'] = [x['sentence_similarity'] for x in most_similar_lines_flat]
    df['sentence_text'] = [x['sentence_text'] for x in most_similar_lines_flat]
    df['abstract_sentence_text'] = [x['abstract_sentence_text'] for x in most_similar_lines_flat]
    df['section_size'] = [x['section_size'] for x in most_similar_lines_flat]
    df['abstract_size'] = [x['abstract_size'] for x in most_similar_lines_flat]
    df['norm_abstract_position'] = [x['abstract_sentence_idx']/x['abstract_size'] for x in most_similar_lines_flat]
    df['norm_position'] = [x['sentence_idx']/x['section_size'] for x in most_similar_lines_flat]
    df['paper_id'] = [x['id'] for x in most_similar_lines_flat]
    
    df['intro_sentence_idx'] = [x['sentence_idx'] if x else None for x in most_similar_intro_lines_flat]
    df['intro_sentence_similarity'] = [x['sentence_similarity'] if x else None for x in most_similar_intro_lines_flat]
    df['intro_sentence_text'] = [x['sentence_text'] if x else None for x in most_similar_intro_lines_flat]
    
    df['conc_sentence_idx'] = [x['sentence_idx'] if x else None for x in most_similar_conc_lines_flat]
    df['conc_sentence_similarity'] = [x['sentence_similarity'] if x else None for x in most_similar_conc_lines_flat]
    df['conc_sentence_text'] = [x['sentence_text'] if x else None for x in most_similar_conc_lines_flat]
    
    return df



def get_most_similar(ds_dict, n=1, is_pubmed = True):
    total_conflated_counts_inst = {}
    total_conflated_counts_abs = {}
    
    keys = list(ds_dict.keys())
    if not is_pubmed:
        keys.remove("1702.05001")
        keys.remove("1103.6094")
            
    for key in keys:
        inst = ds_dict[key]
        
        inst_sections = set()
        
        for ab_sent_dict in inst:
            
            most_similar_sec_dicts = []
            
            for section_sim_dict in ab_sent_dict['section_similarities']:
                
                sorted_line_sims = sorted(section_sim_dict['line_similarities'], reverse=True, key=lambda x: x['sentence_similarity'])
                
                inst_sections.add(conflate_headings(section_sim_dict['section_name']))
               
                
                for i in range(n):
                    if len(list(sorted_line_sims)) > i:
                        most_similar_line_dict = list(sorted_line_sims)[i]
                        most_similar_line_dict['section_name'] = section_sim_dict['section_name']
                        most_similar_line_dict['section_size'] = len(section_sim_dict['line_similarities'])
                        most_similar_line_dict['abstract_sentence_idx'] = ab_sent_dict['sentence_idx']
                        most_similar_line_dict['abstract_sentence_text'] = ab_sent_dict['sentence_text']
                        most_similar_line_dict['abstract_size'] = len(inst)
                        most_similar_line_dict['id'] = key
                        most_similar_sec_dicts.append(most_similar_line_dict)                    
            
            ab_sent_dict["section_top_similarities"] = most_similar_sec_dicts
            
            for i in range(n):
                if len(ab_sent_dict['section_top_similarities']) > i: 
                    ab_sent_dict[f"most_similar_line_{i}"] = list(sorted(ab_sent_dict['section_top_similarities'], reverse=True, key=lambda x: x['sentence_similarity']))[i]
                else:
                    ab_sent_dict[f"most_similar_line_{i}"] = list(sorted(ab_sent_dict['section_top_similarities'], reverse=True, key=lambda x: x['sentence_similarity']))[0]
    
            for sec_sim_dict in ab_sent_dict["section_top_similarities"]:
                if conflate_headings(sec_sim_dict['section_name']) == "[introduction]":
                    ab_sent_dict[f"most_similar_intro_line"] = sec_sim_dict
                    
                if conflate_headings(sec_sim_dict['section_name']) == "[conclusion]":
                    ab_sent_dict[f"most_similar_conc_line"] = sec_sim_dict

        for sec_name in list(inst_sections):
            total_conflated_counts_inst[sec_name] = total_conflated_counts_inst[sec_name] + 1 if \
                sec_name in total_conflated_counts_inst.keys() else 1
                
            total_conflated_counts_abs[sec_name] = total_conflated_counts_abs[sec_name] + len(inst) if \
                sec_name in total_conflated_counts_abs.keys() else len(inst)
            
    return get_most_similar_df(ds_dict, keys, n), total_conflated_counts_inst, total_conflated_counts_abs
                
pubmed_ms_df, pubmed_conf_counts, pubmed_conf_counts_abs = get_most_similar(pubmed_dict, N)        
arxiv_ms_df, arxiv_conf_counts, arxiv_conf_counts_abs = get_most_similar(arxiv_dict, N, False)
```

```python
# This is an issue with this analysis, results are dominated by some papers 
# Is there any way to work around this?
def print_unique_ids(df):
    df["id"] = df.index
    df_section_counts = df.groupby('paper_id')['id'].nunique()
    with pd.option_context('display.min_rows', 30):
        print(df_section_counts.sort_values(ascending=False))    
        print(df['paper_id'].nunique())
        
print_unique_ids(active_df)
```

```python
''' See which sections occur the most as the most similar, and compare these with abstract sentence positions '''

def print_conflated_sections(df):
    df["id"] = df.index
    df_section_counts = df.groupby('section_name_conflated')['id'].nunique()
    with pd.option_context('display.min_rows', 30):
        print(df_section_counts.sort_values(ascending=False))    
        print(df['section_name_conflated'].nunique())
        
    return list(df_section_counts.sort_values(ascending=False).keys())[:6]


active_df['section_name_conflated'] = active_df['section_name']
    
active_df["section_name_conflated"] = active_df["section_name_conflated"].apply(
    conflate_headings
)
    
top_6_conflated = print_conflated_sections(active_df)
```

```python
''' If N is 3, check how often intro/conc are in the top 3 for each abstract sentence '''

counts = []

if N == 3:
    total = 0
    intro_top3_count = 0
    conc_top3_count = 0
    
    # seperate by paper
    paper_ids = active_df['paper_id'].unique()
    
    for pid in list(paper_ids):
        df_pid_insts = active_df[active_df.paper_id.isin([pid])]
    
        sentence_idxs = df_pid_insts['abstract_sentence_idx'].unique()
        

        # We want to iterate through each df_pid_inst and check for abstract or 
        # conclusion in each of the 3 instances that should be present for each 
        count = 0
        for idx in sentence_idxs:
            ab_sent_idx_insts = df_pid_insts[df_pid_insts.abstract_sentence_idx.isin([idx])]
                        
            total += 1
            
            for inst in ab_sent_idx_insts.iterrows():
                if inst[1].section_name_conflated == "[introduction]":
                    intro_top3_count += 1
                    break
            
            for inst in ab_sent_idx_insts.iterrows():
                if inst[1]['section_name_conflated'] == "[conclusion]":
                    conc_top3_count += 1
                    break
    
    print(f"Introduction top 3 occurences: {intro_top3_count}")
    print(f"Conclusion top 3 occurences:   {conc_top3_count}")
    print(f"Total sentences confirmation:  {total}")
            
```

```python
''' Look at Similarity score averages for each conflated section '''

print("Average Similarities")
for section in top_6_conflated:
    section_scores = list(active_df[active_df.section_name_conflated.isin([section])]['sentence_similarity'])
    print(f"{section}" + " "*(20 - len(section)) + \
          f"= {round(np.mean(section_scores), 3)} +- {round(np.std(section_scores), 3)}")
    
```

```python
''' Look at the avg difference between the most similar line and the most similar intro/conc line '''

# Filter out instances with no intro/conclusion sections
intro_idxs = active_df["intro_sentence_text"].apply(lambda x: x != None)
conc_idx = active_df["conc_sentence_text"].apply(lambda x: x != None)

df_filtered_intro = active_df[intro_idxs] 
df_filtered_conc = active_df[conc_idx] 

# Filter out instances where the most similar instance is already from the intro/conclusion
df_filtered_intro = df_filtered_intro[df_filtered_intro.section_name_conflated != "[introduction]"]
df_filtered_conc = df_filtered_conc[df_filtered_conc.section_name_conflated != "[conclusion]"]

df_filtered_intro['intro_sim_diff'] = df_filtered_intro.sentence_similarity - df_filtered_intro.intro_sentence_similarity 
df_filtered_conc['conc_sim_diff'] = df_filtered_conc.sentence_similarity - df_filtered_conc.conc_sentence_similarity 

# Intro differences
print("INTRO")
for section in top_6_conflated:
    df_filtered = df_filtered_intro[df_filtered_intro.section_name_conflated.isin([section])]
    section_scores = list(df_filtered["intro_sim_diff"])
    print(f"{section}" + " "*(20 - len(section)) + \
        f": {round(np.mean(section_scores), 3)} +- {round(np.std(section_scores), 3)} ({len(df_filtered)})")
    
# Conc differences
print("\nCONC")
for section in top_6_conflated:
    df_filtered = df_filtered_conc[df_filtered_conc.section_name_conflated.isin([section])]
    section_scores = list(df_filtered["conc_sim_diff"])
    print(f"{section}" + " "*(20 - len(section)) + \
        f": {round(np.mean(section_scores), 3)} +- {round(np.std(section_scores), 3)} ({len(df_filtered)})")
```

```python
import re, arxiv

unique_doms = set([re.sub("[0-9.]", "", x) for x in arxiv_ids])

dom_counts = {}
for prefix in unique_doms:
    pre = prefix.strip()
    dom_ids = [x for x in arxiv_ids if re.sub("[0-9.]", "", x) == prefix]
    count = len(dom_ids)
    dom_counts[pre] = count
    print(f'{pre} : {count} (e.g. {dom_ids[0].strip()})')
    
```

```python
pre = ""

pre_paper_ids = [x for x in arxiv_ids if re.sub("[0-9.]", "", x) == pre + "\n"]

print(len(pre_paper_ids))

for i, paper_id in enumerate(pre_paper_ids):
    if i%100 == 0:
        print(i)
        print(dom_counts)
    search = arxiv.Search(id_list=[paper_id])
    paper = next(search.results())
    cat = paper.primary_category
    dom_counts[cat] = dom_counts[cat] + 1 if cat in dom_counts else 1 
```
